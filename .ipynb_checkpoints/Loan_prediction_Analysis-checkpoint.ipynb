```python
'''
Now let us look at some of the categorical variable is more details. 
Let us look at the chances o fgetting a loan based on credit history. We will use excel like pivot tables and 
cross-tabulation
Loan status is coded as 1 for yes and 0 for No. So the mean will represent the probability of getting a loan 
'''
```

```python
temp1 = df['Credit_History'].value_counts(ascending=True)
temp2 = df.pivot_table(values = 'Loan_Status',index=['Credit_History'],aggfunc=lambda x:x.map({'Y':1,'N':0}).mean())
print ("Frequency Table for credit history")
print (temp1)

print ("Probability of getting loan for each credit history class:")
print(temp2)
```

```python
# We can use "matplotlib" to plot a bar chart as follows
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(121)
ax1.set_xlabel('Credit History')
ax1.set_ylabel('No. of applicants')
ax1.set_title("Frequency of Applicant by Crediti Hisotry")
temp1.plot(kind='bar')

ax2= fig.add_subplot(122)
ax2.set_xlabel('Credit History')
ax1.set_ylabel('Probability of getting a loan')
ax1.set_title("Probability of getting a loan by credit history")
temp2.plot(kind='bar')
```

```python
# combining the graphs in a stacked chart
temp3 = pd.crosstab(df['Credit_History'],df['Loan_Status'])
temp3.plot(kind = 'bar', stacked=True, color= ['red','blue'], grid=False)
```
